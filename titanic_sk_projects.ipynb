```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('/content/titanic.csv')
data
```

```python
data[['Ticket', 'Cabin']].isnull().sum()
```

```python
if 'Cabin' in data.columns and 'Ticket' in data.columns:
  data.drop(columns=['Cabin', 'Ticket'], inplace=True)
  print(data)
else:
  print("Columns 'Cabin' and/or 'Ticket' have already been dropped.")
```

```python
print(data)
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
data
```

```python
data1 = data['Pclass'].value_counts()
print(data1)
```

# 1. Survival Count by Class with hue for Gender

```python
data['Survived'] = data['Survived'].map({0: 'Not Survived', 1: 'Survived'})

fig = px.histogram(data, x='Pclass', color='Survived', barmode='group',
                   category_orders={"Survived": ['Not Survived', 'Survived']},
                   color_discrete_sequence=px.colors.sequential.Viridis)

fig.update_layout(
    title='Survival Count by Class and Gender',
    xaxis_title='Passenger Class',
    yaxis_title='Count',
    legend_title='Survived',
    bargap=0.2,
    width=1000,
    height=600
)

fig.show()
```

2. Age Distribution by Survival with KDE plot
bold text



```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', hue='Survived', kde=True, palette='viridis', multiple='stack', alpha=0.5)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Survived', labels=['Not Survive', 'Survived'])
plt.show()
```

# 3. Fare Distribution by Class with Swarm plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Pclass', y='Fare', palette='viridis')
sns.swarmplot(data=data, x='Pclass', y='Fare', color=".25")
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.yscale('log')
plt.show()
```

# 4. Pairplot for Multivariate Analysis

```python
sns.pairplot(data, hue='Survived', palette='viridis', diag_kind='kde')
plt.show()
```
